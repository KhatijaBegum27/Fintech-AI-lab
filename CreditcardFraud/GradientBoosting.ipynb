```python
# EDA for Credit Card Fraud Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/creditcard.csv')

print(f"Shape: {df.shape}")
display(df.head())

df.info()
display(df.describe())

missing_values = df.isnull().sum().sum()
duplicate_rows = df.duplicated().sum()
print(f"Missing Values: {missing_values}")
print(f"Duplicate Rows: {duplicate_rows}")

print("Data Types Summary:")
print(df.dtypes.value_counts())

print("\nTarget Variable Distribution:")
class_counts = df['Class'].value_counts()
class_percent = df['Class'].value_counts(normalize=True) * 100
print(pd.concat([class_counts, class_percent], axis=1, keys=['Count', 'Percentage']))

plt.figure(figsize=(5,4))
sns.countplot(x='Class', data=df, palette='coolwarm')
plt.title("Fraud vs Non-Fraud Distribution")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Count")
plt.show()

```

```python
# Data Cleaning and Preprocessing #

# 1. Remove duplicate rows
df = df.drop_duplicates()
print("Duplicates removed. \nNew shape:""", df.shape)
print('\n')

# 2.  Outlier Detection
# Boxplot for 'Amount'
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Amount'])
plt.title("Boxplot of Transaction Amounts")
plt.show()
# Cap Outliers
cap = df['Amount'].quantile(0.99)
df['Amount'] = np.where(df['Amount'] > cap, cap, df['Amount'])
print("Capped extreme outliers in 'Amount' at 99th percentile.\n")

# 3. Correlation Heatmap
plt.figure(figsize=(8,4))
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
corr_target = corr['Class'].sort_values(ascending=False)
print("\nTop correlated features with Class:")
print(corr_target.head(10))
print('\n')
```

```python
# 4. Feature Scaling (Only on Amount & Time)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df['scaled_amount'] = scaler.fit_transform(df[['Amount']])
df['scaled_time'] = scaler.fit_transform(df[['Time']])

df = df.drop(['Amount', 'Time'], axis=1)
print("Scaled 'Amount' and 'Time' and dropped originals.")
```

```python
# Split Train & Test

from sklearn.model_selection import train_test_split

X = df.drop('Class', axis=1)
y = df['Class']

# Stratified split keeps same fraud ratio in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Train distribution:\n", y_train.value_counts(normalize=True))
print("\nTest distribution:\n", y_test.value_counts(normalize=True))
```

```python
## Model Training → Evaluation ##

import xgboost as xgb

scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()  # imbalance ratio

xgb_model = xgb.XGBClassifier(
    tree_method='hist',
    n_estimators=300,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight,  # handles imbalance
    eval_metric='auc',
    random_state=42
)
xgb_model.fit(X_train, y_train)

y_pred_xgb = xgb_model.predict(X_test)
y_pred_xgb_prob = xgb_model.predict_proba(X_test)[:, 1]
```

```python
import lightgbm as lgb

lgb_model = lgb.LGBMClassifier(
    n_estimators=300,
    learning_rate=0.05,
    num_leaves=64,
    subsample=0.8,
    colsample_bytree=0.8,
    class_weight='balanced',   # handles imbalance automatically
    random_state=42
)
lgb_model.fit(X_train, y_train)

y_pred_lgb = lgb_model.predict(X_test)
y_pred_lgb_prob = lgb_model.predict_proba(X_test)[:, 1]
```

```python
# Model Evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

def evaluate_model(y_true, y_pred, y_prob, title):
    auc = roc_auc_score(y_true, y_prob)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    print(f"\n📊 {title}")
    print(f"AUC: {auc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1-Score: {f1:.4f}")

    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {title}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

evaluate_model(y_test, y_pred_xgb, y_pred_xgb_prob, "XGBoost (Class Weight)")
evaluate_model(y_test, y_pred_lgb, y_pred_lgb_prob, "LightGBM (Class Weight)")
```

```python
# XGBoost Threshold Tuning & Re-evaluating Model
from sklearn.metrics import precision_recall_curve, roc_auc_score, f1_score, precision_score, recall_score, confusion_matrix

precisions, recalls, thresholds = precision_recall_curve(y_test, y_pred_xgb_prob)
f1_scores = 2 * (precisions * recalls) / (precisions + recalls)

# Maximize F1
best_f1_idx = np.argmax(f1_scores)
best_f1_threshold = thresholds[best_f1_idx]
print(f" Best Threshold (Max F1): {best_f1_threshold:.3f}")
print(f"Precision: {precisions[best_f1_idx]:.4f}, Recall: {recalls[best_f1_idx]:.4f}, F1: {f1_scores[best_f1_idx]:.4f}")

# Maximize Recall with Precision>0.85
valid_idxs = np.where(precisions >= 0.85)[0]
if len(valid_idxs) > 0:
    best_recall_idx = valid_idxs[np.argmax(recalls[valid_idxs])]
    best_recall_threshold = thresholds[best_recall_idx]
    print(f"\n Best Threshold (Max Recall, Precision≥0.85): {best_recall_threshold:.3f}")
    print(f"Precision: {precisions[best_recall_idx]:.4f}, Recall: {recalls[best_recall_idx]:.4f}")
else:
   print("\n No threshold found where Precision ≥ 0.85")

y_pred_f1 = (y_pred_xgb_prob >= best_f1_threshold).astype(int)
y_pred_recall = (y_pred_xgb_prob >= best_recall_threshold).astype(int)

def model_summary(y_true, y_pred, y_prob, title):
    auc = roc_auc_score(y_true, y_prob)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    print(f"\n {title}")
    print(f"AUC: {auc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")

model_summary(y_test, y_pred_f1, y_pred_xgb_prob, "XGBoost (Tuned for Max F1)")
model_summary(y_test, y_pred_recall, y_pred_xgb_prob, "XGBoost (Tuned for High Recall, Precision≥0.85)")
```

```python
# Plot Precision–Recall–F1 vs Threshold
plt.figure(figsize=(8,6))
plt.plot(thresholds, precisions[:-1], label='Precision', color='blue')
plt.plot(thresholds, recalls[:-1], label='Recall', color='green')
plt.plot(thresholds, f1_scores[:-1], label='F1-Score', color='red')

plt.axvline(best_f1_threshold, color='purple', linestyle='--', label=f'Max F1 = {best_f1_threshold:.3f}')
plt.axvline(best_recall_threshold, color='orange', linestyle='--', label=f'Max Recall (≥0.85 Precision) = {best_recall_threshold:.3f}')

plt.title('Precision–Recall–F1 vs Probability Threshold (XGBoost)')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Confusion Matrix (High Recall)
cm_recall = confusion_matrix(y_test, y_pred_recall)
sns.heatmap(cm_recall, annot=True, fmt='d', cmap='Blues')
plt.title(f'Confusion Matrix - XGBoost (High Recall, Threshold={best_recall_threshold:.3f})')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
