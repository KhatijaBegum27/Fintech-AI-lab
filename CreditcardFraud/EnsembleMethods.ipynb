```python
# EDA for Credit Card Fraud Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/creditcard.csv')

print(f"Shape: {df.shape}")
display(df.head())

df.info()
display(df.describe())

missing_values = df.isnull().sum().sum()
duplicate_rows = df.duplicated().sum()
print(f"Missing Values: {missing_values}")
print(f"Duplicate Rows: {duplicate_rows}")

print("Data Types Summary:")
print(df.dtypes.value_counts())

print("\nTarget Variable Distribution:")
class_counts = df['Class'].value_counts()
class_percent = df['Class'].value_counts(normalize=True) * 100
print(pd.concat([class_counts, class_percent], axis=1, keys=['Count', 'Percentage']))

plt.figure(figsize=(5,4))
sns.countplot(x='Class', data=df, palette='coolwarm')
plt.title("Fraud vs Non-Fraud Distribution")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Count")
plt.show()

```

```python
# Data Cleaning and Preprocessing #

# 1. Remove duplicate rows
df = df.drop_duplicates()
print("Duplicates removed. \nNew shape:""", df.shape)
print('\n')

# 2.  Outlier Detection
# Boxplot for 'Amount'
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Amount'])
plt.title("Boxplot of Transaction Amounts")
plt.show()
# Cap Outliers
cap = df['Amount'].quantile(0.99)
df['Amount'] = np.where(df['Amount'] > cap, cap, df['Amount'])
print("Capped extreme outliers in 'Amount' at 99th percentile.\n")

# 3. Correlation Heatmap
plt.figure(figsize=(8,4))
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
corr_target = corr['Class'].sort_values(ascending=False)
print("\nTop correlated features with Class:")
print(corr_target.head(10))
print('\n')
```

```python
# 4. Feature Scaling (Only on Amount & Time)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df['scaled_amount'] = scaler.fit_transform(df[['Amount']])
df['scaled_time'] = scaler.fit_transform(df[['Time']])

df = df.drop(['Amount', 'Time'], axis=1)
print("Scaled 'Amount' and 'Time' and dropped originals.")
```

```python
# 5. Handle Class Imbalance (SMOTE)

# Split Train & Test

from sklearn.model_selection import train_test_split

X = df.drop('Class', axis=1)
y = df['Class']

# Stratified split keeps same fraud ratio in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Train distribution:\n", y_train.value_counts(normalize=True))
print("\nTest distribution:\n", y_test.value_counts(normalize=True))

# Apply SMOTE

from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("\nBefore SMOTE:\n", y_train.value_counts())
print("\nAfter SMOTE:\n", y_train_res.value_counts())

# Verify shapes
print("Train (balanced):", X_train_res.shape, "Test:", X_test.shape)
```

```python
## Model Training â†’ Evaluation ##

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve

# Logistic Regression
log_reg = LogisticRegression(
    max_iter=300,
    solver='lbfgs',
    class_weight='balanced',
    n_jobs=-1,
    random_state=42
)
log_reg.fit(X_train_res, y_train_res)
y_pred_lr = log_reg.predict(X_test)
y_pred_lr_prob = log_reg.predict_proba(X_test)[:,1]

# Random Forest
rf = RandomForestClassifier(
    n_estimators=60,
    max_depth=10,
    min_samples_split=10,
    class_weight='balanced',
    n_jobs=-1,
    random_state=42
)
rf.fit(X_train_res, y_train_res)
y_pred_rf = rf.predict(X_test)
y_pred_rf_prob = rf.predict_proba(X_test)[:,1]

# Model Evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(y_true, y_pred, y_prob, title):
    auc = roc_auc_score(y_true, y_prob)
    print(f"\n{title}")
    print("AUC:", round(auc, 4))
    print("Precision:", round(precision_score(y_true, y_pred), 4))
    print("Recall:", round(recall_score(y_true, y_pred), 4))
    print("F1-Score:", round(f1_score(y_true, y_pred), 4))
    sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

evaluate_model(y_test, y_pred_lr, y_pred_lr_prob, "Logistic Regression")
evaluate_model(y_test, y_pred_rf, y_pred_rf_prob, "Random Forest")
```
